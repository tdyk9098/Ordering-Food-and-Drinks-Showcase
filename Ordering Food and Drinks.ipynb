```python
#Example 1
# Menu dictionary shows the possible order configurations
menu = {"fries": {"small": "$4", "medium": "$6.25", "large": "$7.75"},
          "drink": {"small": "$2", "medium": "$4.5", "large": "$5.75"},
          "burger": {"veggies": "$8.35", "cheese": "$8.55", "combo": "$9.55"}}
# Option list shows the keywords for ordering a menu item 
option = ["yes","no"]
# Cost is an empty list to be appended with the cost total
cost = []

# Ordering fries
while True:
    print("Would you like fries?(Yes/No)\n")
    fries = input()
    if option[0] in fries.lower() or option[1] in fries.lower():
        while "yes" in fries.lower():
            print("\nWould you like small, medium, or large fries?\n")
            fries_size = input()
            if fries_size.lower() == "small":
                cost.append(eval(list(list(menu.values())[0].values())[0][1:]))
                break
            elif fries_size.lower() =="medium":
                cost.append(eval(list(list(menu.values())[0].values())[1][1:]))
                break
            elif fries_size.lower()== "large":
                cost.append(eval(list(list(menu.values())[0].values())[2][1:]))
                break
            else:
                print("\nI'm sorry, I didn't understand that.")
        break
    else:
        print("\nI'm sorry, I didn't understand that.")
        continue

#Ordering a drink
while True:
    print("\nWould you like a drink?(Yes/No)\n")
    drink = input()
    if option[0] in drink.lower() or option[1] in drink.lower():
        while "yes" in drink.lower():
            print("\nWould you like a small, medium, or large drink?\n")
            drink_size = input()
            if drink_size.lower() == "small":
                cost.append(eval(list(list(menu.values())[1].values())[0][1:]))
                break
            elif drink_size.lower() == "medium":
                cost.append(eval(list(list(menu.values())[1].values())[1][1:]))
                break
            elif drink_size.lower()== "large":
                cost.append(eval(list(list(menu.values())[1].values())[2][1:]))
                break
            else:
                print("\nI'm sorry, I didn't understand that.")
        break
    else:
        print("\nI'm sorry, I didn't understand that.")
        continue

# Ordering a burger
while True:
    print("\nWould you like a burger?(Yes/No)\n")
    burger = input()
    if option[0] in burger.lower() or option[1] in burger.lower():
        while "yes" in burger.lower():
            print("\nWould you like a veggie, cheese, or combo burger?\n")
            burger_style = input()
            if burger_style.lower() == "veggie":
                cost.append(eval(list(list(menu.values())[2].values())[0][1:]))
                break
            elif burger_style.lower() == "cheese":
                cost.append(eval(list(list(menu.values())[2].values())[1][1:]))
                break
            elif burger_style.lower() == "combo":
                cost.append(eval(list(list(menu.values())[2].values())[2][1:]))
                break
            else:
                print("\nI'm sorry, I didn't understand that.")
        break
    else:
        print("\nI'm sorry, I didn't understand that.")
        continue

# Printing the total cost        
print("\n\nGreat! Your total is $"+str(sum(cost)))
```
